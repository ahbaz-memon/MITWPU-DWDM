# Exercise 02

### A.Importing the header Files

```python
import numpy as np #for effiemathematical 
import matplotlib.pyplot as plt #for visualisation
from sklearn import model_selection #for splitting dataset
from sklearn import linear_model # linear Regeression
```

### B.Loading datasets(Weight vs Height)

```python
path="C:/Users/SALMAN MEMON/Desktop/weight-height.csv"

data=np.loadtxt(path,delimiter=',')
data=np.array(data[0:100])
```

```python
data.shape
```

```python
X=data[:,0]
Y=data[:,1]
X.shape,Y.shape
```

```python
#visualization of X and Y

plt.scatter(X,Y,c='r',alpha=0.8)
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### C.Splitting dataset into training and testing

```python
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,random_state=11,test_size=0.3)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

### D.Visualising Traing points and testing points

```python
#visualization of X_train and Y_train

plt.scatter(X_train,Y_train,c='g',alpha=0.5)
plt.title('X_train vs Y_train')
plt.xlabel('X_train')
plt.ylabel('Y_train')
plt.show()
```

```python
#visualization of X_test and Y_test

plt.scatter(X_test,Y_test,c='c',alpha=0.5)
plt.title('X_test vs Y_test')
plt.xlabel('X_test')
plt.ylabel('Y_test')
plt.show()
```

### E.fitting the taining data for Linear Regression

```python
LR=linear_model.LinearRegression()
```

```python
X_train=X_train.reshape((-1,1))
Y_train=Y_train.reshape((-1,1))

LR.fit(X_train,Y_train)
```

### F.Predicting the testing data point

```python
X_test=X_test.reshape((-1,1))
Y_test=Y_test.reshape((-1,1))
Y_pred=LR.predict(X_test)
```

### G.Info about best fit line

```python
m=LR.coef_[0]
c=LR.intercept_[0]
print("slope of best fit line : ",m)
print("intercept of best fit line : ",c)

print("score of training points : ",LR.score(X_train,Y_train))
print("score of testing points : ",LR.score(X_test,Y_test))

a=np.arange(0,10)
plt.plot(a,((m*a)+c),color='k')
plt.show()
```

### H.Visualization of traing and testing points with best fit line

```python
#visualization of X_train and Y_train

plt.scatter(X_train,Y_train,c='g',alpha=0.5)
plt.plot(X_train,((X_train*m)+c),color='k')
plt.title('X_train vs Y_train')
plt.xlabel('X_train')
plt.ylabel('Y_train')
plt.show()
```

```python
#visualization of X_test and Y_test

plt.scatter(X_test,Y_test,c='c',alpha=0.5)
plt.plot(X_test,((X_test*m)+c),color='k')
plt.title('X_test vs Y_test')
plt.xlabel('X_test')
plt.ylabel('Y_test')
plt.show()
```
