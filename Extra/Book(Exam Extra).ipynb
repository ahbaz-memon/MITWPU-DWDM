# Libraries

```python
import time # for noting times for algorithm
import numpy as np # for efficient list
import pandas as pd # for efficient tables 
from matplotlib import pyplot as plt # for visualization 
import seaborn as sns # for visualization
from mlxtend.frequent_patterns import fpgrowth # machine learning xtend tool for fp growth algorithm
```

# Loading data.....

```python
path  = "Book.txt" # path at my PC
f = open(path, 'r') # declaring file pointer and openning file in read mode only
```

# Preprocessing starts...

```python
data1 = [] # empty list for store collection of element of list
for line in f.readlines(): # reading line by line in every iteration
    temp = [] # empty list for store element of list
    for e in line.split(','): # iterating every element of lines in splitted ','
        if e != '' and e != '\n': # Avaiding for ' '(spaces) and '\n'(new-line)
            temp.append(e.strip()) # adding in temp
    data1.append(temp) # adding collection of temp's
```

```python
columns = set([]) # empty set since set skip the same entries
for l in data1: # a list in data
    for e in l: # a element in list
        columns.add(e)
columns = list(columns) # convert set to list to easy iterate
columns = sorted(columns) # sort elements by alphabetical order
columns
```

```python
bool_data1 = []
for l in data1: # a list in data
    temp = []
    for c in columns: # every column in columns
        if c in l: # if column is in the list
            temp.append(True) # append the value True 
        else:
            temp.append(False) # append the value False
    bool_data1.append(temp) 
```

```python
data = pd.DataFrame(bool_data1, columns = columns) # creating data frame
```
