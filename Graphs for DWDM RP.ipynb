```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
x1 = pd.read_csv('d01.csv').drop(['Unnamed: 0'], axis = 1)
x1 = x1.iloc[:, 0].to_list()
```

```python
x2 = x1.copy()
for i in range(len(x2)):
    if x2[i] > 292:
        x2[i] = -150
    x2[i] -= 25    
```

```python
plt.figure(figsize = (14, 8))
plt.xticks(range(1, len(x1) + 1, 2))
plt.yticks([])
plt.xlabel('Question number', fontsize = 15)
plt.title('Behavior of Null entries', fontsize = 20)
plt.grid()
plt.plot(range(1, len(x1) + 1), np.array(x1) + 200, 'g', alpha = 0.3, linewidth = 4,label = 'stage-1')
plt.scatter(range(1, len(x1) + 1), np.array(x1) + 200, c = 'k', alpha = 0.65)

plt.plot(range(1, len(x2) + 1), np.array(x2) + 75, 'b', alpha = 0.3, linewidth = 4,label = 'stage-2')
plt.scatter(range(1, len(x2) + 1), np.array(x2) + 75, c = 'k', alpha = 0.65)

plt.plot(range(1, len(x1) + 1), np.array(len(x1) * [-100]), 'm', alpha = 0.3, linewidth = 4,label = 'stage-3')
plt.scatter(range(1, len(x1) + 1), len(x1) * [-100], c = 'k', alpha = 0.65)
plt.legend()
#plt.show()
plt.savefig('g01.jpg',bbox_inches = 'tight', facecolor='white')
```

```python
data = pd.read_csv("03.Cleaned Mental Health.csv").drop(['Unnamed: 0'], axis = 1)
data
```

```python
# plt.figure(figsize = (100,100))
fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1, )
fig.set_figheight(12)
fig.set_figwidth(15)
plt.subplots_adjust(hspace = 0.3)

sns.countplot(data = data, x = 'country_live', hue = 'tech_comp_flag', alpha = 0.9, ax = ax1)
ax1.set_title('No. of Respondents for Country Live and tech_flag', fontsize = 20)
ax1.set_xticklabels(ax1.get_xticklabels())
ax1.set_xlabel('Country Live', fontsize = 18)
ax1.set_ylabel('Count', fontsize = 18)
ax1.legend(['Not in Tech', 'In Tech'])

sns.countplot(data = data, x = 'country_work', hue = 'tech_comp_flag', alpha = 0.9, ax = ax2)
ax2.set_title('No. of Respondents for Country Work and tech_flag', fontsize = 20)
ax2.set_xticklabels(ax2.get_xticklabels())
ax2.set_xlabel('Country Work', fontsize = 18)
ax2.set_ylabel('Count', fontsize = 18)
ax2.legend(['Not in Tech', 'In Tech'])

#plt.show()
fig.savefig('g02.jpg',bbox_inches = 'tight', facecolor='white')
```
