```python
def species_to_color(k):
    temp = []
    for e in k:
        if e == 'Iris-setosa':
            temp.append('r')
        elif e == 'Iris-virginica':
            temp.append('b')
        elif e == 'Iris-versicolor':
            temp.append('g')
    return temp
```

```python
fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(111,projection = '3d')
x = data.iloc[:,0].to_list()
y = data.iloc[:,1].to_list()
z = data.iloc[:,2].to_list()
ax.scatter(x, y, z, c = 'r')
```

# Split the data

```python
X = data.iloc[:,0:4]
Y = data.iloc[:,4]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 2, test_size = 0.2)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

# Creating Obj of DecisionTreeClassifier

```python
DT = tree.DecisionTreeClassifier(max_depth=2)
```

```python
DT.fit(X_train, Y_train)
```

# predicting

```python
Y_pred = DT.predict(X_test)
```

# Report

```python
print("score of training :",metrics.accuracy_score(Y_train,DT.predict(X_train)))
print("score of testing :",metrics.accuracy_score(Y_test,Y_pred))
```

```python
print("Classification report :\n",classification_report(Y_test,Y_pred))
```
